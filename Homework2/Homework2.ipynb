```python
import numpy as np
def inter(x, y):
    common_tokens = set(y).intersection(x)
    if len(common_tokens) == 0:
        distance = np.nan
    else:
        distance = 1/len(common_tokens)
    return distance
```

```python
def get_k_neighbours(k, dataframe, reviews_train_data):
    dataframe = dataframe.sort_values(by=['distance'])
    dataframe = dataframe[dataframe['distance'].rank(method='dense')<=k]
    new_dataframe = pd.DataFrame(columns=['k_nearest_train_text','label'])
    for row in dataframe.itertuples():
        new_dataframe = new_dataframe.append({'k_nearest_train_text': row.train_text, 'label': reviews_train_data['label'][reviews_train_data['reviews_text'] == row.train_text].values[0]}, ignore_index=True)
    return new_dataframe
```

```python
predicted_test_data = pd.DataFrame(columns=['test_reviews_text','predicted_label'])
i = 0
for test_row in reviews_test_data.itertuples():
    test_row_tokens = test_row.reviews_text.split(' ')
    string_match = pd.DataFrame(columns=['train_text','distance'])
    for train_row in reviews_train_data.itertuples():
        train_text = train_row.reviews_text
        train_row_tokens = train_row.reviews_text.split(' ')
        string_match = string_match.append({'train_text': train_text, 'distance': inter(train_row_tokens, test_row_tokens)}, ignore_index=True)
    string_match = get_k_neighbours(2,string_match,reviews_train_data)
    predicted_num0 = string_match['label'].value_counts()['0']
    predicted_num1 = string_match['label'].value_counts()['1']
    if predicted_num0 > predicted_num1:
        predicted_test_data = predicted_test_data.append({'test_reviews_text': test_row.reviews_text, 'predicted_label': 0}, ignore_index=True)
    else:
        predicted_test_data = predicted_test_data.append({'test_reviews_text': test_row.reviews_text, 'predicted_label': 1}, ignore_index=True)
    print('i is : ',i)
    print('predicted_test_data : ',predicted_test_data)
    i = i + 1
    if i == 100:
        break
#print(string_match)
```
