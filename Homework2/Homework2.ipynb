```python
import numpy as np
def inter(x, y):
    common_tokens = set(y).intersection(x)
    if len(common_tokens) == 0:
        distance = np.nan
    else:
        distance = 1/len(common_tokens)
    return distance
```

```python
def get_k_neighbours(k, dataframe):
    dataframe = dataframe.sort_values(by=['distance'])
    dataframe = dataframe[dataframe['distance'].rank(method='dense')<=k]
    #print(dataframe)
    new_dataframe = pd.DataFrame(columns=['train_text','label'])
    for row in dataframe.itertuples():
        new_dataframe = new_dataframe.append({'train_text': row.train_text, 'label': 0}, ignore_index=True)
    #print(new_dataframe)    
    return new_dataframe
```

```python
for test_row in reviews_test_data.itertuples():
    test_row = test_row.reviews_text.split(' ')
    string_match = pd.DataFrame(columns=['train_text','distance'])
    for train_row in reviews_train_data.itertuples():
        train_text = train_row.reviews_text
        train_row = train_row.reviews_text.split(' ')
        string_match = string_match.append({'train_text': train_text, 'distance': inter(train_row, test_row)}, ignore_index=True)
        string_match = get_k_neighbours(2,string_match)
    #print(string_match)
    break
#print(string_match)
```
